```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Загрузка данных (предполагается, что у вас есть файл train.csv с колонками 'text' и 'target')
data = pd.read_csv('IMDB Dataset.csv')

# Разделение данных на тренировочный и тестовый наборы
X_train, X_test, y_train, y_test = train_test_split(data['review'],data['sentiment'], test_size=0.2, random_state=42)

# Векторизация текста с использованием TF-IDF
vectorizer = TfidfVectorizer(max_features=6000)  # можно изменить максимальное количество признаков
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# Обучение модели логистической регрессии
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)

# Предсказание на тестовом наборе
predictions = model.predict(X_test_tfidf)

# Оценка модели
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')


```
