### Что после редизайна время использования личного кабинета на сайте изменилось
$H_0$: Время после редезайна `==` как и до редезайна <br>
$H_1$: Время после редезайна `>` чем до редезайна

```python
alpha = 0.05

t_statics, p_value = st.ttest_rel (
    time_before,time_after
)


print(f'p-value: {p_value}')

if p_value < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```
